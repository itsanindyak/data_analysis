# **Bag of word**

```python
import pandas as pd
import numpy as np
```

```python
df = pd.DataFrame({"text":["people watch dswithbappy",
                         "dswithbappy watch dswithbappy",
                         "people write comment",
                          "dswithbappy write comment"],"output":[1,1,0,0]})

df
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv= CountVectorizer()
```

```python
bow = cv.fit_transform(df['text'])
cv.vocabulary_ # it give index to every unique word
```

```python
bow.toarray()
```

```python
bow[0].toarray() # it give array for 0 index means first sentence
```
